# Task 1 :- IRIS FLOWER CLASSIFICATION

```python
# Importing necessary library

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# We'll load the iris dataset using the load_iris() function from Scikit-learn.

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
# Visualizing the dataset using scatter plots for better understanding

plt.figure(figsize=(10, 7))
plt.scatter(X[:, 0], X[:, 1], c=y, label="Species")
plt.xlabel("Sepal length (cm)")
plt.ylabel("Sepal width (cm)")
plt.legend()
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Using K-Neighbors Algorithm to classify the Iris Flowers 

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
